```python
files = {
    'actor.csv' : {'prefix':'act'},
    'address.csv' : {'prefix':'add'},
    'category.csv': {'prefix':'cat'},
    'city.csv': {'prefix':'city'},
    'country.csv':{'prefix':'ct'},
    'customer.csv':{'prefix':'cust'},
    'film.csv':{'prefix':'film'},
    'film_actor.csv':{'prefix':'fa'},
    'film_category.csv':{'prefix':'fc'},
    'inventory.csv':{'prefix':'inv'},
    'language.csv':{'prefix':'lang'},
    'payment.csv':{'prefix':'pay'},
    'rental.csv':{'prefix':'rent'},
    'staff.csv':{'prefix':'staff'},
    'store.csv':{'prefix':'store'},

}
```

```python
len(files)
```

```python
# create graph reduce nodes
gr_nodes = {
    f.split('/')[-1]: DynamicNode(
        fpath=f,
        fmt='csv',
        pk='id',
        prefix=files[f]['prefix'],
        date_key=None,
        compute_layer=GraphReduceComputeLayerEnum.pandas,
        compute_period_val=730,
        compute_period_unit=PeriodUnit.day,
    )
    for f in files.keys()
}
```

```python
gr = GraphReduce(
    name='cust_dynamic_graph',
    parent_node=gr_nodes['customer.csv'],
    fmt='csv',
    cut_date=datetime.datetime(2023,9,1),
    compute_layer=GraphReduceComputeLayerEnum.pandas,
    auto_features=True,
    auto_feature_hops_front=1,
    auto_feature_hops_back=2,
    #label_node=gr_nodes['orders.csv'],
    #label_operation='count',
    #label_field='id',
    #label_period_val=60,
    #label_period_unit=PeriodUnit.day
)
```

```python
for ix, row in labels.iterrows():
    gr.add_entity_edge(
        parent_node=gr_nodes[row['to_name']],
        relation_node=gr_nodes[row['from_name']],
        parent_key=row['to_key'],
        relation_key=row['from_key'],
        reduce=True
    )
```

```python
gr.do_transformations()
```

```python
gr = GraphReduce(
    name='rental_dynamic_graph',
    parent_node=gr_nodes['customer.csv'],
    fmt='csv',
    cut_date=datetime.datetime(2022,6,1),
    compute_layer=GraphReduceComputeLayerEnum.pandas,
    compute_period_val=365,
    compute_period_unit=PeriodUnit.day,
    auto_features=True,
    auto_feature_hops_front=1,
    auto_feature_hops_back=2,
    label_node=gr_nodes['payment.csv'],
    label_operation='sum',
    label_field='amount',
    label_period_val=30,
    label_period_unit=PeriodUnit.day
)
```

```python
for ix, row in labels.iterrows():
    gr.add_entity_edge(
        parent_node=gr_nodes[row['to_name']],
        relation_node=gr_nodes[row['from_name']],
        parent_key=row['to_key'],
        relation_key=row['from_key'],
        reduce=True
    )
```

```python
for node in gr.nodes():
    node.reload()
```

```python
# need to tell the payment.csv what date to use
gr_nodes['payment.csv'].date_key = 'payment_date'
```

```python
gr.do_transformations()
```
